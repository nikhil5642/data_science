```python
import pandas as pd
import numpy as np
```

```python
train=pd.DataFrame.from_csv('train.tsv',sep='\t')
train
```

```python
test=pd.DataFrame.from_csv('test.tsv',sep='\t')
test
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
l_state=LabelEncoder()
l_state.fit(train.Phrase)
state_enc=(list(l_state.classes_))
state_enc=np.array(state_enc)
#df_state=pd.DataFrame(state_enc)
#print(df_state)
train.Phrase = l_state.transform(train.Phrase)
train.info()
```

```python
from sklearn import ensemble

clf = ensemble.RandomForestClassifier(n_estimators=200,n_jobs=-1,random_state=0)
clf.fit(train.Phrase, train.Sentiment)

prediction=clf.predict(test['Phrase'])
prediction
```
