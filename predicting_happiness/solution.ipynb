```python
def ignore_(word):
    word=word.lower()
    code=["ignor","rude"]
    if any(x in word for x in code):
          return 1
    return 0
```

```python
def river_(word):
    word=word.lower()
    if "river" in word:
          return 1
    return 0
```

```python
def bad_(word):
    word=word.lower()
    code=["good","like","love","best","nice"]
    if any(x in word for x in code):
            return 1
    return 0
   
```

```python
def night_(word):
    word=word.lower()
    if "night" in word:
        return 1
    return 0
```

```python
def smoking(word):
    word=word.lower()
    if "smok" in word:
        return 1
    return 0
```

```python
def price_(word):
    word=word.lower()
    code=["cost","price","rate","charge"]
    if any(x in word for x in code):
        if "low" in word:
            return 0
        elif "high" in word:
            return 2
    return 1
     
```

```python
def comfort_(word):
    word=str(word.lower())
    if "comfort" in word:
        return 1
    return 0  
```

```python
def child_(word):
    word=word.lower()
    code=["son","daughter","child","kid"]
    if any(x in word for x in code):
        return 1
    return 0 
```

```python
def room_(word):
    word=word.lower()
    code=["clean","nice","perfect"]
    if any(x in word for x in code):
        return 1
    return 0 
```

```python
def relation_(word):
    word=word.lower()
    code=["boyfriend","girlfriend","wife","husband","hubby"]
    if any(x in word for x in code):
        return 1
    return 0 
```

```python
def staff_(word):
    word=word.lower()
    code=["best","friend","help","excellent"]
    if "staff" in word:
        if any(x in word for x in code):
             return 1
    return 0 
```

```python
def service_(word):
    word=word.lower()
    code=["great","good","nice"]
    if "serv" in word:
        if any(x in word for x in code):
             return 1
    return 0 
```
