```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
train['Description'][466]
```

```python
def smoking(word):
    word=word.lower()
    if "smok" in word:
        return 1
    return 0
```

```python
def price_(word):
    word=word.lower()
    code=["cost","price","rate"]
    if any(x in word for x in code):
        if "low" in word:
            return 0
        elif "high" in word:
            return 2
    return 1
     
```

```python
def comfort_(word):
    word=str(word.lower())
    if "comfort" in word:
        return 1
    return 0  
```

```python
def child_(word):
    word=word.lower()
    code=["son","daughter","child","kid"]
    if any(x in word for x in code):
        return 1
    return 0 
```

```python
def room_(word):
    word=word.lower()
    code=["clean","nice","perfect"]
    if any(x in word for x in code):
        return 1
    return 0 
```

```python
def relation_(word):
    word=word.lower()
    code=["boyfriend","girlfriend","wife","husband","hubby"]
    if any(x in word for x in code):
        return 1
    return 0 
```

```python
def staff_(word):
    word=word.lower()
    code=["best","friend","help"]
    if "staff" in word:
        if any(x in word for x in code):
             return 1
    return 0 
```

```python
def service_(word):
    word=word.lower()
    code=["great","good","nice"]
    if "serv" in word:
        if any(x in word for x in code):
             return 1
    return 0 
```

```python
for df in train,test:
    smoke,price,comfort,child,room,relation,staff,service=[],[],[],[],[],[],[],[]
    for i in range(len(df)):
        smoke.append(smoking(df['Description'][i]))
        comfort.append(comfort_(df['Description'][i]))
        price.append(price_(df['Description'][i]))
        child.append(child_(df['Description'][i]))
        room.append(room_(df['Description'][i]))
        relation.append(relation_(df['Description'][i]))
        staff.append(staff_(df['Description'][i]))
        service.append(service_(df['Description'][i]))

    df['smoke']=pd.DataFrame(smoke) 
    df['comfort']=pd.DataFrame(comfort) 
    df['price']=pd.DataFrame(price) 
    df['child']=pd.DataFrame(child) 
    df['room']=pd.DataFrame(room) 
    df['relation']=pd.DataFrame(relation) 
    df['staff']=pd.DataFrame(staff) 
    df['service']=pd.DataFrame(service) 
```

```python
browser=['Chrome','Edge','Firefox','Google Chrome','IE','Internet Explorer','InternetExplorer','Mozilla','Mozilla Firefox','Opera','Safari']
```

```python
for df in train,test:
    for key, value in zip(browser,range(len(browser))):
        df.loc[df['Browser_Used'] == key, "Browser_Used"] = value
    
train 
```

```python
response=['not happy','happy']
for df in train,test:
    for key, value in zip(response,(0,1)):
        df.loc[df['Is_Response'] == key, 'Is_Response'] = value
```

```python
device=['Desktop','Mobile','Tablet']
for df in train,test:
    for key, value in zip(device,range(len(device))):
        df.loc[df['Device_Used'] == key, "Device_Used"] = value
    
train 
```

```python
train=train.drop(['User_ID','Description'],axis=1)
test=test.drop(['Description'],axis=1)
```

```python
train=pd.DataFrame(train)
train
```

```python
from sklearn.ensemble import RandomForestClassifier
x_train=train.drop('Is_Response', axis=1)
y_train=train['Is_Response']
clf = RandomForestClassifier(n_estimators=200,n_jobs=-1, random_state=1)
clf.fit(x_train,y_train )
prediction = clf.predict(test.drop("User_ID", axis=1))

submission=test["User_ID"].append(prediction,"Is_Response")
submission.to_csv("submission.csv", index=False)
```
