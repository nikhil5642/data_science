```python
import pandas as pd
import numpy as np
import re
```

```python
train=pd.read_csv('train.csv')
train
```

```python
test=pd.read_csv('test.csv')
test
```

```python
train.info()
test.info()
```

```python
import calendar
from datetime import datetime
for dailyData in train,test:
    dailyData["date"] = dailyData.datetime.apply(lambda x : x.split()[0])
    dailyData["hour"] = dailyData.datetime.apply(lambda x : x.split()[1].split(":")[0])
    dailyData["weekday"] = dailyData.date.apply(lambda dateString : calendar.day_name[datetime.strptime(dateString,"%Y-%m-%d").weekday()])
    dailyData["month"] = dailyData.date.apply(lambda dateString : calendar.month_name[datetime.strptime(dateString,"%Y-%m-%d").month])

```

```python
train=train.drop("datetime",axis=1)
train.head()
```

```python
test.head()
```

```python
train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l_hour=LabelEncoder()
l_hour.fit(train['hour'])
train['hour']=l_hour.transform(train['hour'])
test['hour']=l_hour.transform(test['hour'])
```

```python
l_week=LabelEncoder()
l_week.fit(train['weekday'])
train['weekday']=l_week.transform(train['weekday'])
test['weekday']=l_week.transform(test['weekday'])
```

```python
l_month=LabelEncoder()
l_month.fit(train['month'])
train['month']=l_month.transform(train['month'])
test['month']=l_month.transform(test['month'])
    
```

```python
train=train.drop(['date','casual','registered'],axis=1)
```

```python
test=test.drop(['date'],axis=1)
test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier()
clf.fit(train.drop(['count'],axis=1),train['count'])
```

```python
prediction=clf.predict(test.drop(['datetime'],axis=1))
prediction
```

```python
submission=pd.DataFrame({'datetime':test['datetime']})
submission['count']=pd.DataFrame(prediction)
submission    
```

```python
submission.to_csv('submission2.csv',index=False)
```
