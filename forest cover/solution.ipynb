```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('train.csv')
train
```

```python
test=pd.read_csv('test.csv')
test
```

```python
train=train.drop(['Id'],axis=1)
train
```

```python
train.info()
```

```python
train.describe()
```

```python
train.groupby('Cover_Type').size()
```

```python
x_train=train.drop(['Cover_Type'],axis=1)
x_train
```

```python
y_train=train['Cover_Type']
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=200,n_jobs=-1,random_state=0)
clf.fit(x_train,y_train)
prediction = clf.predict(test.drop("Id", axis=1))
prediction
 
```

```python
from sklearn.ensemble import ExtraTreesClassifier
clf = ExtraTreesClassifier(n_estimators=100,n_jobs=-1,random_state=0)
clf.fit(x_train,y_train)
prediction2 = clf.predict(test.drop("Id", axis=1))
prediction2
    
```

```python
submission = pd.DataFrame({"Id": test["Id"]})
submission['Cover_Type']=pd.DataFrame(prediction,columns=['Cover_Type'])
submission.to_csv('submission.csv',index=False)
```

```python
submission2 = pd.DataFrame({"Id": test["Id"]})
submission2['Cover_Type']=pd.DataFrame(prediction2,columns=['Cover_Type'])
submission2.to_csv('submission.csv',index=False)
```
